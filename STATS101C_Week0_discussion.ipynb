We want a 5 by 5 matrix with ones on the outer element, 0 on the inner elements and 9 in the center element


```python
output = np.ones((5,5))
print(output)

z = np.zeros((3,3))
z[1,1] = 9
print(z)

output[1:-1,1:-1] = z
print(output)
```
